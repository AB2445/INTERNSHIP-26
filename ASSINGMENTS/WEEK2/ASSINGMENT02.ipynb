```python
"""
1.Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with
word_count, char_count (excluding spaces), unique_words (case-insensitive), and
most_frequent_word. Test it with
""";
```

```python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    word_count = len(words)
    char_count = len(sentence.replace(" ", ""))
    unique_words = len(set(words))

    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1

    most_frequent_word = max(freq, key=freq.get)

    return {
        "word_count": word_count,
        "char_count": char_count,
        "unique_words": unique_words,
        "most_frequent_word": most_frequent_word
    }


```

```python
analyze_sentence("The quick brown fox jumps over the lazy dog the fox")
```

```python
"""
2. Given the string below, do the following and print each result:
text = "A man a plan a canal Panama"
a) Reverse the string without using slicing.
b) Check if it is a palindrome, ignoring case and spaces.
c) Replace every vowel with *.
d) Count how many times each vowel appears
""";
```

```python
text = "A man a plan a canal Panama"

# a) Reverse without slicing
reverse_text = ""
for ch in text:
    reverse_text = ch + reverse_text

print("Reversed String:", reverse_text)

# b) Palindrome check (ignore case and spaces)
clean_text = text.replace(" ", "").lower()

if clean_text == "".join(reversed(clean_text)):
    print("Palindrome: Yes")
else:
    print("Palindrome: No")

# c) Replace vowels with *
vowels = "aeiouAEIOU"
replaced_text = ""

for ch in text:
    if ch in vowels:
        replaced_text += "*"
    else:
        replaced_text += ch

print("After Replacing Vowels:", replaced_text)

# d) Count each vowel
vowel_count = {'a':0, 'e':0, 'i':0, 'o':0, 'u':0}

for ch in text.lower():
    if ch in vowel_count:
        vowel_count[ch] += 1

print("Vowel Counts:", vowel_count)

```

```python
""" 3. Write a function format_report(data) that takes a list of tuples (name, score, grade) and prints a
formatted table where name is left-aligned (width 15), score is right-aligned (width 6), and grade is
centered (width 8). Use this data:
data = [("Alice", 92, "A"), ("Bob", 78, "B+"), ("Charlie", 55, "C"), ("Diana",
100, "A+")]
""";
```

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")
    print("-" * 29)

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")


data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
""""   4. Write a function list_ops(nums) that:
a) Removes duplicates while preserving the original order.
b) Returns a new list with only elements at even indices, squared.
c) Flattens the nested list into a single list using list comprehension.
nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5] nested = [[1, 2, 3], [4, 5], [6, 7, 8,
9]]
""";
```

```python
def list_ops(nums, nested):

    # a) Remove duplicates while preserving order
    unique_nums = []
    for num in nums:
        if num not in unique_nums:
            unique_nums.append(num)

    # b) Elements at even indices, squared
    even_index_squared = [nums[i] ** 2 for i in range(0, len(nums), 2)]

    # c) Flatten nested list using list comprehension
    flattened = [item for sublist in nested for item in sublist]

    return unique_nums, even_index_squared, flattened


nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

unique_nums, even_index_squared, flattened = list_ops(nums, nested)

print("After Removing Duplicates:", unique_nums)
print("Even Index Elements Squared:", even_index_squared)
print("Flattened List:", flattened)
```
